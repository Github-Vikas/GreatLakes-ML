```python
# Now moving onto CLustering , we find that maximum number of peaks is observed in the column age and hence we wil be taking the value 
## for peaks from (2,6 ) ; dropping the column strentgh as it is to be used for predi


con_fe=con_norm.drop(columns='strength')
```

```python
from scipy.stats import zscore
```

```python
con_fez = con_fe.apply(zscore)
```

```python
con_fez ## MAtrix of z score
```

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
clusters = range(2,6)
meanDistortions = []

for k in clusters:
    model= KMeans(n_clusters=k)
    model.fit(con_fez)
    prediction= model.predict(con_fez)
    meanDistortions.append(sum(np.min(cdist(con_fez,model.cluster_centers_,'euclidean'),axis =1))/con_fez.shape[0])
    
plt.plot(clusters,meanDistortions,'bx-')
plt.xlabel('k')
plt.ylabel('Average distortion')
plt.title('Selecting k with elbow method')
```

```python
# Here we don't see any elbow as per the plot which means there are no cluster formation in the dataset itself and hence need to deop the idea of finding the k cluster.
con_fez.head()
```

```python
cov_matrix = np.cov(con_fez.T)
```

```python
print('Covariance matrix \n%s',cov_matrix)
```

```python
eigenvalues,eigenvectors = np.linalg.eig(cov_matrix)

print("\nEigen vectors \n%s ",eigenvectors)
print("\nEigen Values \n%s",eigenvalues)
```

```python
eig_pairs  = [(eigenvalues[index],eigenvectors[:,index]) for index in range (len(eigenvalues))]
eig_pairs.sort()
eig_pairs.reverse()
print(eig_pairs)
```

```python
eigenvalues_sorted = [eig_pairs[index][0] for index in range(len(eigenvalues))]
eigenvectors_sorted = [eig_pairs[index][1] for index in range(len(eigenvalues))]



print('\n Eigenvalues in descending order \n %s' %eigenvalues_sorted)
```

```python
tot = sum(eigenvalues)
var_explained = [(i/tot) for i in sorted(eigenvalues, reverse=True)]
cum_var_exp = np.cumsum(var_explained)
```

```python
cum_var_exp
```

```python
plt.bar(range(1,9),var_explained,alpha=0.5, align= 'center',label='individual explained variance')
plt.step(range(1,9),cum_var_exp,where= 'mid',label = 'cumulative explained variance')
plt.ylabel('Explained Variance ratio')
plt.xlabel('Principal components')
plt.legend(loc='best')
plt.show()
```

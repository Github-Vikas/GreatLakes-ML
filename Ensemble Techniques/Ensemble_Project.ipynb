```python
## COming to the importing for relevant libraries for model building. 

from sklearn.model_selection import train_test_split

from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier

from sklearn import metrics

```

```python
# Changing the name of the target column to label
new.rename(columns = {"y":"label"},inplace= True)
```

```python
# Splitting data into train and test set with 70 : 30 ratio. 


X=new.drop("label" , axis=1 )
y=new['label']
test_size = 0.30
seed = 7
X_train,X_test,y_train, y_test = train_test_split(X,y,test_size=test_size,random_state=seed)
```

```python
# Calling the decision tree classifer for fitting in the model . 

dt_model = DecisionTreeClassifier()
dt_model.fit(X_train,y_train)
```

```python
# predicting the score
y_predict = dt_model.predict(X_test)
print(dt_model.score(X_train,y_train))
print(dt_model.score(X_test,y_test))
```

```python
# Confusion matrix for accuracy of the model
from sklearn.metrics import confusion_matrix
confusion_matrix(y_true=y_test,y_pred =y_predict)
```

```python
# metrics classification report (f1 -score)

print(metrics.classification_report(y_test,y_predict))
```

```python
# trying to improve the model using the ensemble techniques

rfc_1 = RandomForestClassifier(n_estimators = 50 )
rfc_1= rfc_1.fit(X_train, y_train)

y_predict = rfc_1.predict(X_test)
print(rfc_1.score(X_train, y_train))
print(rfc_1.score(X_test, y_test))
```

```python
# Confusion matrix for the random forest. 
from sklearn.metrics import confusion_matrix
print("Confusion MAtrix : \n")

print(pd.DataFrame(confusion_matrix(y_test,y_predict,labels=[0,1])))
```

```python
print(metrics.classification_report(y_test,y_predict))
```

```python
# using the gradient boosting  classifier for building an ensemble of model. 


from sklearn.ensemble import GradientBoostingClassifier
gbc1 = GradientBoostingClassifier(n_estimators = 50,learning_rate = 0.12,max_depth = 5)
gbc1 = gbc1.fit(X_train,y_train)
```

```python
y_predict = gbc1.predict(X_test)

print(gbc1.score(X_train, y_train))
print(gbc1.score(X_test, y_test))
```

```python
print("Confusion MAtrix :\n")

print(pd.DataFrame(confusion_matrix(y_test,y_predict,labels= [0,1])))
```

```python
print(metrics.classification_report(y_test, y_predict))
```
